```python
import requests
import pandas as pd
```

```python
## Pull from Chicago Data Portal

# Documentation 1 (API usage): http://dev.socrata.com/foundry/#/data.cityofchicago.org/ucdv-yd74
# Documentation 2 (Paging): http://dev.socrata.com/docs/paging.html

# API settings
endpoint = 'https://data.cityofchicago.org/resource/ucdv-yd74.json'
page_size = 50000 #maximum page size

# Counters and holders
offset = 0
request_counter = 0
data = []
responses = ['start'] #Start off with len(responses) > 0

# Get data
while len(responses) > 0 and request_counter < 10: #limit request_counter so it doesn't go forever
    r = requests.get('{0}?$limit={1}&$offset={2}'\
                     .format(endpoint, page_size, offset))
    responses = r.json()
    request_counter += 1
    offset = offset + page_size
    data.extend(responses)

# Diagnostics
print 'Number of requests: {0}'.format(request_counter)
print 'Number of rows retrieved: {0}'.format(len(data))

# Transform into Pandas dataframe
chicago_data = pd.DataFrame(data)
chicago_data
```

```python
## Pull from Census

# Documentation: http://www.census.gov/data/developers/data-sets/acs-survey-5-year-data.html

# API settings
endpoint = 'http://api.census.gov/data/2013/acs5/profile'
census_api_key = 'e495b05c8bdb52b20fd696e8a906d7d12e5e7f01'
variables = 'NAME,DP03_0062E' # Get Name of area and variable for median household income
state = 17 #Illinois

# Get data
request_url = '{0}?get={1}&for=tract:*&in=state:{2}+county:*&key={3}'\
              .format(endpoint, variables, state, census_api_key)
r = requests.get(request_url)

# Prep to transform into a Pandas dataframe
request_list = r.json()
request_headers = request_list.pop(0)

# Transform into Pandas dataframe
census_data = pd.DataFrame(request_list, columns=request_headers)
census_data
```
